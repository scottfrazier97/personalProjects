### Sorting

```python
#you can sort a list in place by calling it's sort function. Sort() assigns None to the list, so sort first and then resassign.
list1 = [5,3,2,6,7,4]
list1.sort()
sorted_list = list1
print(sorted_list)
```

```python
#secondary sort key - a function that produces a value to sort the objects. F.e.: Can sort based on lengths of objects
new_list = ["saw", "small", "He", "foxes", "six"]
new_list.sort(key=len)
new_list
```

### Slicing

```python
#you can select sections of most sequence types by using slice notation.
#typical form is: [start: end: step]
#uses 0 indexing. Starts at index 2 (which is the third item in list) and ends with index 4 (which is fifth item in list, but 
#in this case, it does not include ending item)
new_list[2:4]
```

```python
#to reverse a list, do the following
new_list[::-1]
```
