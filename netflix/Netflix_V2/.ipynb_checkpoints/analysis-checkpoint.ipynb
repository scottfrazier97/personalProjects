```python
df[['category','language']] = df.category.str.split(expand=True) #Splitting category column into two separate columns, one is entertainment category and the other is language
df
```

```python
df = df.rename(columns={'week': 'date'})
```

```python
df
```

```python
df[['year','month','day']] = df.date.str.split('-',expand=True) #Splitting category column into two separate columns, one is entertainment category and the other is language
df
```

```python
df.drop(["date"], axis=1)
```

```python
# Concatenate show_title and season_title, replacing NaN values in season_title with show_title
df['title'] = df['season_title'].fillna(df['show_title'])

# Drop the original show_title and season_title columns
df.drop(['show_title', 'season_title'], axis=1, inplace=True)

# Display the DataFrame with the new 'title' column
df["category"]=="TV".head()
```
