# iloc --- indexing

```python
#Bring back specific value based on row, and column POSITION using iloc(INDEXING),
#regular loc brings back value based on STRINGS

#iloc is not used widely, because you typically will not know the exact position of what you want
dfLoc = netflixDF.iloc[3,4]
dfLoc
```

```python
#Same thing as above, but quicker
netflixDF.iloc[3,4]
```

```python
#Pulling back multiple columns for specific row using iloc
netflixDF.iloc[3, 3:5]
```

# loc --- strings, booleans

```python
netflixDF.head()
```

```python
#Pull back all rows using : in the row position within brackets, and the new cols list with desired columns
cols = ["director", "rating"]

netflixDF.loc[:, cols]
```

```python
#Another way of doing the above ^^^
netflixDF.loc[:, ["director", "rating"]]
```

# using a mask

```python
#Pulling specific rows based on a conditional
netflixDF["country"] == "United States"
```

```python
#Pulling specific rows based on a conditional,
#colon in the loc is optional, brings back all rows by default
mask = netflixDF["country"] == "United States"
netflixDF.loc[mask, :]
```

```python
#list comprehension + mask
mask2 = ["Movie" in x for x in netflixDF["type"]]
netflixDF.loc[mask2]
```
